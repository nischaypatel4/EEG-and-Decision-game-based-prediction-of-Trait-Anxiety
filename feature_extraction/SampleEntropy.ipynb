```python
import os
import mne
import numpy as np
import pandas as pd
from antropy import sample_entropy
import matplotlib.pyplot as plt
import seaborn as sns
```

**Overview of Sample Entropy:**

Sample Entropy (SampEn) is a non-linear measure of the complexity or irregularity of a time-series signal. It tells you how predictable a signal is — a higher SampEn means more complexity (less predictability), while lower SampEn indicates a more regular (predictable) signal.
It's widely used in EEG, ECG, HRV, and other biosignals, especially for mental state classification, sleep studies, seizure detection, etc.

**Mathematical Intuition**
Let’s go step by step.
***1. Parameters***

    
    r: Tolerance (usually a fraction of the signal’s std, e.g., 0.2 * std)

    m: Embedding dimension (length of sequences to compare, typically 2)

    N: Length of the time series


***2. Concept:***

  We want to find how often patterns of length m repeat, and whether they still match when extended to length m+1.
  Count pairs of sequences of length m that are similar (within distance r)


Count how many of those are still similar when extended to length m+1


Sample Entropy is the negative log of the ratio of these counts.


***3. Formula***
Let:

    A= number of matches of length m+1


    B= number of matches of length m


Then:

    SampEn(m,r,N) = − ln⁡(A/B)
    If a signal is highly predictable, extending the match length doesn’t reduce the count much, so SampEn is low.
    If it's unpredictable, extending reduces matches a lot → SampEn is high.


```python
import matplotlib.pyplot as plt
import seaborn as sns# === Sample Entropy Params ===
m = 2
r_factor = 0.2  # optional if you normalize

# === Loop through all .set files ===
for filename in os.listdir(input_dir):
    if filename.endswith(".set"):
        base = filename.replace(".set", "")
        print(f" Processing {filename}...")

        # Load epoched EEG data
        filepath = os.path.join(input_dir, filename)
        epochs = mne.read_epochs_eeglab(filepath)

        # Data: (epochs, channels, time)
        data = epochs.get_data()
        n_epochs, n_channels, n_times = data.shape

        # Init output array
        sampen_values = np.zeros((n_channels, n_epochs))

        # Compute entropy
        for ch in range(n_channels):
            for ep in range(n_epochs):
                signal = data[ep, ch, :]
                sampen_values[ch, ep] = sample_entropy(signal, order=m)

        # Mean entropy per channel
        mean_sampen = np.mean(sampen_values, axis=1)

        # Save results
        pd.DataFrame(sampen_values.T).to_csv(
            os.path.join(output_dir, f"{base}_sampen_values.csv"), index=False
        )
        pd.DataFrame(mean_sampen).to_csv(
            os.path.join(output_dir, f"{base}_mean_sampen.csv"),
            index=False,
            header=["Mean_SampEn"]
        )

        print(f"Done with {base}")

print(" All files processed and saved to EEG_Entropy_Results!")
```
